```python
import pandas as pd
```

### FPL Data Cleaning

```python
fpl_levels = pd.read_csv('../data/administrative_info/original/fpl.csv')
fpl_levels
```

```python
fpl_levels.columns = [col.replace('\n', ' ') for col in fpl_levels.columns]
fpl_levels = fpl_levels[['Year', "First Person", "Each Additional Person"]]
fpl_levels.head()
```

```python
fpl_levels.loc[8, 'Each Additional Person'] = '$4,170'
```

```python
fpl_levels.loc[:, 'Each Additional Person'] = fpl_levels['Each Additional Person'].str.replace(r'\D+', '', regex=True).astype(int)
fpl_levels.loc[:, 'First Person'] = fpl_levels['First Person'].str.replace(r'\D+', '', regex=True).astype(int)
fpl_levels
```

```python
fpl_levels.to_csv("../data/administrative_info/cleaned/fpl.csv")
```

### Standard Deduction Cleaning

```python
sd = pd.read_csv('../data/administrative_info/original/standard_deductions.csv', delim_whitespace=True, header = None)
sd
```

```python
sd.columns = ['Year', 'Single Person', 'Head of Household', 'Married Couple']
sd
```

```python
sd.loc[:, 'Single Person'] = sd['Single Person'].str.replace(r'\D+', '', regex=True).astype(int)
sd.loc[:, 'Head of Household'] = sd['Head of Household'].str.replace(r'\D+', '', regex=True).astype(int)
sd.loc[:, 'Married Couple'] = sd['Married Couple'].str.replace(r'\D+', '', regex=True).astype(int)
sd
```

```python
sd.to_csv("../data/administrative_info/cleaned/standard_deductions.csv")
```
