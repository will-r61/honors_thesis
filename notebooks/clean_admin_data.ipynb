```python
import pandas as pd
```

```python
fpl_levels = pd.read_csv('../data/administrative_info/original/fpl.csv')
fpl_levels
```

```python
fpl_levels.columns = [col.replace('\n', ' ') for col in fpl_levels.columns]
fpl_levels = fpl_levels[['Year', "First Person", "Each Additional Person"]]
fpl_levels.head()
```

```python
fpl_levels.loc[8, 'Each Additional Person'] = '$4,170'
```

```python
fpl_levels.loc[:, 'Each Additional Person'] = fpl_levels['Each Additional Person'].str.replace(r'\D+', '', regex=True).astype(int)
fpl_levels.loc[:, 'First Person'] = fpl_levels['First Person'].str.replace(r'\D+', '', regex=True).astype(int)
fpl_levels
```

```python
fpl_levels.to_csv("../data/administrative_info/cleaned/fpl.csv")
```
