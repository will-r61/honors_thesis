```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = np.arange(0, 20)

state_b = ts / 40 + 0.25 + np.random.beta(1, 99, 20)
# State A and B are not equal, but are trending similarly
state_a = state_b + np.random.beta(1, 99, 20) + 0.1
# Get state A without treatment
state_a_no_treatment_post = state_a[9:].copy()
# Add in treament effect
state_a[10:] = state_a[10:] - np.random.beta(20, 80, 10)
```

```python
plt.plot(ts, state_a, linestyle = '-', label = "State A", color = 'black')
plt.plot(ts[9:], state_a_no_treatment_post, linestyle = '--', label = "State A Without Treatment", color = 'black')
plt.plot(ts, state_b, linestyle = '-', label = "State B", color = 'lightcoral')
plt.axvline(x = 9, label = "Treatment Time", linestyle = ':', lw = 2)
plt.xlabel("t")
plt.ylabel("Y")
plt.title("Parallel Trends")
plt.legend()
plt.savefig('../output/synthetic_control_plots/parallel_trends.png')
```
